```python
import numpy as np
import matplotlib.pyplot as plt
num=50
np.random.seed(0)
x=np.r_[np.random.randn(num,2)-[1,1],np.random.randn(num,2)+[1,1]]
y=[-1]*num+[1]*num
plt.figure()
for i in range(len(x)):
    if y[i]==1:
        plt.plot(x[i][0],x[i][1],'ro')
    else:
        plt.plot(x[i][0],x[i][1],'bo')
```

```python
w=np.array([-1,1])
b=0
delta=1e-3

for i in range(len(x)):
    if y[i]==1:
        plt.plot(x[i][0],x[i][1],'ro')
    else:
        plt.plot(x[i][0],x[i][1],'bo')
        
for i in range(101):
    grad_x=np.array([0,0])
    grad_b=0
    for j in range(len(x)):
        if y[j]!=np.sign(np.dot(w,x[j])+b):
            grad_x=grad_x+y[j]*x[j]
            grad_b=grad_b+y[j]
    w=w+delta*grad_x
    b=b+delta*grad_b
    
    if i%10==0:
        xx=[-4,4]
        yy=[0,0]
        for i in range(len(xx)):
            yy[i]=-(w[0]*xx[i]+b)/w[1]
        pp=plt.plot(xx,yy)

```
